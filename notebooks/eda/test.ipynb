```python
# Import the prediction function and dependencies
import sys
sys.path.append('.')
from netflix_prediction_function import load_and_predict
import pandas as pd
import numpy as np

# Single Prediction Example
# Create sample data for single prediction
single_data = pd.DataFrame({
    'lag_1': [0.02],
    'lag_2': [0.01],
    'lag_3': [-0.005],
    'moving_avg_5': [0.015],
    'moving_avg_10': [0.012],
    'volatility_10': [0.008]
})

# Make single prediction
try:
    prediction, metadata = load_and_predict('netflix_stock_rf_model.pkl', 
                                           'netflix_stock_rf_metadata.json', 
                                           single_data)
    print(f"Single Prediction: {prediction[0]:.6f}")
    print(f"Model Metadata: {metadata}")
except Exception as e:
    print(f"Error: {e}")
    print("Make sure model files exist in the correct directory")
```

```python
# Batch Prediction Example
# Create sample data for batch prediction (multiple rows)
batch_data = pd.DataFrame({
    'lag_1': [0.02, 0.015, -0.01, 0.008, 0.025],
    'lag_2': [0.01, 0.02, 0.015, -0.005, 0.012],
    'lag_3': [-0.005, 0.01, 0.02, 0.015, -0.008],
    'moving_avg_5': [0.015, 0.018, 0.012, 0.020, 0.014],
    'moving_avg_10': [0.012, 0.015, 0.018, 0.011, 0.016],
    'volatility_10': [0.008, 0.010, 0.007, 0.009, 0.011]
})

print(f"Batch data shape: {batch_data.shape}")
print("\nFirst few rows of batch data:")
print(batch_data.head())

# Make batch predictions
try:
    predictions, metadata = load_and_predict('netflix_stock_rf_model.pkl', 
                                           'netflix_stock_rf_metadata.json', 
                                           batch_data)
    print(f"\nBatch Predictions: {predictions}")
    print(f"Number of predictions: {len(predictions)}")
    print(f"Average prediction: {np.mean(predictions):.6f}")
    print(f"Prediction std: {np.std(predictions):.6f}")
except Exception as e:
    print(f"Error: {e}")
    print("Make sure model files exist in the correct directory")
```
